## **Devoluciones**

Se decidió eliminar las columnas de FECHA_DOC y FECHA_ENT, además de la columna de SERIE que tenía un mismo valor. También se eliminó la columna DES_FIN que tenía todos los valores en 0

```python
df_devoluciones=df_devoluciones.drop(["FECHA_DOC","FECHA_ENT","SERIE","DES_FIN"],axis=1)
df_devoluciones
```

```python
df_devoluciones.info()
```

Se asigna a los valores nulos de la columna vendedor el 0 como identificador de aquellas devoluciones de las cuales no se tiene registro de vendedor

```python
df_devoluciones['VENDEDOR'] = df_devoluciones['VENDEDOR'].fillna(0)
```

Se eliminarán aquellos registros nulos de la columna CVE_PEDI. La condicion es que si esta vacia la fila, busque en la columna de CVE_DOC el formato F12345 (Ejemplo). 

```python
#PASAMOS LOS VALORES DE CVE_DOC A CVE_PEDI SI LA CELDA ES NULA
nulos = df_devoluciones['CVE_PEDI'].isnull()
patron = 'F\d{5}'

coincidencias = df_devoluciones.loc[nulos, 'CVE_DOC'].str.contains(patron)

df_devoluciones.loc[nulos & coincidencias, 'CVE_PEDI'] = df_devoluciones.loc[nulos & coincidencias, 'CVE_DOC']
df_devoluciones.info()
```

```python
#PASAMOS LOS VALORES DE CVE_PEDI A CVE_DOC SI LA CELDA ES NULA
nulos2 = df_devoluciones['CVE_DOC'].isnull()
patron2 = 'F\d{5}'

coincidencias2 = df_devoluciones.loc[nulos2, 'CVE_PEDI'].str.contains(patron2)

df_devoluciones.loc[nulos2 & coincidencias2, 'CVE_DOC'] = df_devoluciones.loc[nulos2 & coincidencias2, 'CVE_PEDI']
df_devoluciones.info()
```

```python
df_devoluciones['CVE_DOC'] = df_devoluciones['CVE_DOC'].fillna('SI0000')
df_devoluciones.info()
```

```python
df_devolucionesfiltro = df_devoluciones[df_devoluciones['FECHA_CANCELA'].notnull()]
```

```python
df_devoluciones = df_devoluciones.drop("FECHA_CANCELA",axis=1)
df_devoluciones
```

## **Notas de credito**

```python
#Checar si hay valores nulos por columna
valores_nulos=df_notascredito.isnull().sum()
valores_nulos
```

```python
#Filtro de fechas canceladas 
filtro_cancelaciones = df_notascredito[df_notascredito['FECHA_CANCELA'].notnull()]
filtro_cancelaciones.head()
```

```python
#eliminar columna FECHA_CANCELA del data frame original
df_notascredito = df_notascredito.drop(columns=['FECHA_CANCELA'])
df_notascredito.head()
```

```python
#Cambiar valores nulos en la columna VENDEDOR por 0 (cada vendedor tiene un numero)
df_notascredito['VENDEDOR'].fillna(0, inplace=True)
```

```python
#SI0000 = Valores en la columna CVE_PEDI que estan en blanco
df_notascredito['CVE_PEDI'].fillna('SI0000', inplace=True)
```

```python
#Checar si hay valores nulos por columna
valores_nulos=df_notascredito.isnull().sum()
valores_nulos
```

## **Clientes**

```python
#Checar si hay valores nulos por columna
valores_nulos=df_clientes.isnull().sum()
valores_nulos
```

```python
#SI0000 = Valores en la columna CVE_PEDI que estan en blanco
df_clientes['RFC'].fillna('000000000000', inplace=True)
```
